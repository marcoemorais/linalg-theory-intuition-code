## 2.2 Vectors: geometry and algebra
* vector is a line with a direction and magnitude
* vector has no implicit origin which distinguishes the concept from coordinates which have zero as origin
    * vector $v = \begin{bmatrix}1 \\ 2 \end{bmatrix}$ could be located anywhere in the 2d plane
* negating a vector is same as rotating by $180^o$ (see also Vector-scalar multiplication)

```python
origin = np.array([0, 0])
v = np.array([1, 2])
plt.quiver(*origin, *v, angles='xy', scale_units='xy', scale=1)
plt.axis([-3,3,-3,3])
plt.title('2d Vector Plot with $v$ at origin')
plt.ylabel('$v_2$')
plt.xlabel('$v_1$')
plt.grid()
plt.show()
```

## 2.5 Vector-scalar multiplications

```python
origin = np.array([0, 0])
v = np.array([1, 2])
plt.quiver(*origin, *v, angles='xy', scale_units='xy', scale=1, color='red', label='$v$')
plt.quiver(*origin, *(0.5*v), angles='xy', scale_units='xy', scale=1, color='blue', label='$.5v$')
plt.quiver(*origin, *(-v), angles='xy', scale_units='xy', scale=1, color='green', label='$-v$')
plt.axis([-3,3,-3,3])
plt.title('Scaling Vector $v$')
plt.ylabel('$v_2$')
plt.xlabel('$v_1$')
plt.legend()
plt.grid()
plt.show()
```

## 2.8 Code Challenges

> 1. Create a 2d vector v and 10 scalars drawn at random from a Gaussian normal distribution. Plot the vector v and the 10 vectors formed from the scalar-vector multiplication of v with the 10 scalars. What do you notice?

Answer: All the vectors fall along the same line.

```python
v = np.array([1,1])
s = 2 * np.random.normal(size=10)

# Scale both elements of vector v by the same value drawn from a Gaussian normal.
sv = (np.tile(v, 10) * np.repeat(s, 2)).reshape(10, 2)

axis = 6 * np.array([-1,1,-1,1])

origin = np.array([0, 0])
cmap = plt.cm.Set1
for ind in range(sv.shape[0]):
    plt.quiver(*origin, *sv[ind,:], angles='xy', scale_units='xy', scale=1, 
               color=cmap(ind/10), label='$v_{0}$'.format(ind))
plt.quiver(*origin, *v, angles='xy', scale_units='xy', scale=1, color='black', label='$v$')
plt.axis(axis)
plt.title('Scaling Vector $v$')
plt.ylabel('$v_2$')
plt.xlabel('$v_1$')
plt.legend()
plt.grid()
plt.show()
```
